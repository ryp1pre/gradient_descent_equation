```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Define the function and its derivative:

```python
def fn(x):
    return x ** 2 - 6 * x + 9
```

```python
def fn_derivative(x):
    return 2 * x - 6
```

#### Gradient descent function definition:

```python
def gradient_descent(x, alpha, slope_treshold):
    """
    Perform gradient descent optimization.

    Parameters:
        x (float): Initial value of x.
        alpha (float): Learning rate.
        slope_threshold (float): Threshold for the gradient magnitude to stop optimization.

    Returns:
        tuple: A tuple containing the final x value and its history.
    """
    
    x_history = [x]
    gradient = fn_derivative(x)
    
    while np.abs(gradient) > slope_treshold:
        gradient = fn_derivative(x)
        x = x - alpha * gradient
        x_history.append(x)
        
    return x, x_history
            
```

#### Set parameters and run gradient descent:

```python
x_0 = 14               # initial value
alpha = 0.01           # learning rate
slope_treshold = 0.5   # slope treshold used to stop the optimisation algorithm 

x_final, x_history = gradient_descent(x_0, alpha, slope_treshold)
```

#### Plot the function and optimisation path:

```python
x_axis = np.linspace(-10, 16, 100)

plt.plot(x_axis, fn(x_axis), label = 'Function plot')
plt.scatter(x_history, fn(np.array(x_history)), color = 'red',marker='x', label = 'Optimisation path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent')
plt.legend()
plt.grid()
plt.show()
```
